### Rango

```python
for i in df.columns[8:10]:
  print('Rango', i, ':', max(df[i]) - min(df[i]))
```

```python
for i in df.columns[8:10]:
  for j in df.TERMINAL.unique():
    print("Rango", i, ":", max(df[i][(df.TERMINAL==j)]) - min(df[i][(df.TERMINAL==j)]), "para", j)
```

### Varianza y desviación estándar

```python
for i in df.columns[8:10]:
  print('Varianza', i, ':', df[i].var())

print('-------------------------------')

for i in df.columns[8:10]:
  print('Desviación estándar', i, ':', df[i].std())
```

```python
print("VARIANZA\n")
for i in df.columns[8:10]:
  for j in df.CLASE_VEHICULO.unique():
    print("Varianza", i, ":", df[i][(df.CLASE_VEHICULO==j)].var(), "para", j)

print("-----------------------------------------------------------------")

print("DESVIACIÓN ESTÁNDAR\n")
for i in df.columns[8:10]:
  for j in df.CLASE_VEHICULO.unique():
    print("Desviación estándar", i, ":", df[i][(df.CLASE_VEHICULO==j)].std(), "para", j)
```

```python
import statistics
def aad(x):
   a = mean(absolute(x - statistics.mean(x)))
   return a

for i in df.columns[8:10]:
  print('Desviación Media Absoluta', i, ':', aad(df[i]))
```

## 3.2. Resumen final

Se pudo identificar a partir del anáisis exploratorio que las terminales de transporte en Colombia realizan aproximadamente 2 despachos y movilizan aproximadamente 14 pasajeros. Los niveles de servicio que se identifican a nivel nacional están representados por "Básico" y "Lujo" con 69.6% y 30.4% repectivamente. Esto deja en evidencia que los usuarios de servicios de transporte de pasajeros por carretera no priorizan en un servicio "premium" en este tipo de medios de transporte.

En Bogotá, capital del país, se tiene que los usuarios prefieren únicamente el nivel de servicio "Lujo", lo que puede ser un indicador de mejora del servicio y poder vender servicios de nivle "Preferencial de Lujo". En esta misma ciudad los usuarios tienen suelen utilizar con más frecuencia el "Microbus" y no vehículos más grandes. 

Para las dos variables cuanatitativas del set de datos, Despachos y Pasajeros, se tiene que la cantidad pasajeros presentan una desviación estándar que advierte más variabilidad en cuanto a la demanda de tranporte por carretera. Entre estas dos variables también se tiene una correlación linela positiva de 0.584882, lo que indica que existe un crecimiento en esta proporción entre variables.

Se observó que la clase de vehículo que más pasajeros transporta en promedio es el "Bus", seguido del "Microbus" y "Buseta", pensando de esta manera que invertir en flotas que consideren esta clase de vehículos sería más beneficioso que hacerlo en otras clases.

De manera general en el territorio nacional se tiene que las los horarios de despacho por nivel de servicio se mueven en el rango entre las 8 y 16 horas, con excepción del nivel de servicio "Preferencial de Lujo" que tiene un rango más amplio entre las 7 y 19 horas. Estos rangos horarios facilitan la gestión de la flota disponible de manera que su utilización sea la óptima y no afectar los niveles de satisfacción del cliente.
